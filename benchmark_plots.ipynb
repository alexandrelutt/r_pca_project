```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ATT_performances = dict()

ATT_performances['NMI'] = {'Raw data':0.8563039573116744,
                           'PCA':0.8399617334031217,
                           'RPCA':0.8507415936944136,
                           'GLPCA':0.46396688969087196,
                           'RGLPCA':0.42983684904613495,
                           'Proposed PCA':1.0
                           }


ATT_performances['ARI'] = {'Raw data':0.5869091769612755,
                           'PCA':0.5520067367408009,
                           'RPCA':0.6068177553025976,
                           'GLPCA':0.6068177553025976,
                           'RGLPCA':0.6068177553025976,
                           'Proposed PCA':1.0
                           }

# Plot
plt.figure(figsize=(12, 6))

# NMI Plot
plt.subplot(121)
bar_colors = plt.cm.rainbow(np.linspace(0, 1, len(ATT_performances['NMI'])))
bars = plt.bar(range(len(ATT_performances['NMI'])),
               list(ATT_performances['NMI'].values()),
               align='center',
               color=bar_colors,
               alpha=0.8)
plt.xticks(range(len(ATT_performances['NMI'])),
           list(ATT_performances['NMI'].keys()),
           rotation=45)
plt.title('NMI')
plt.ylim([0, 1])
plt.grid(axis='y', linestyle='--', alpha=0.7)

# ARI Plot
plt.subplot(122)
bar_colors = plt.cm.rainbow(np.linspace(0, 1, len(ATT_performances['ARI'])))
bars = plt.bar(range(len(ATT_performances['ARI'])),
               list(ATT_performances['ARI'].values()),
               align='center',
               color=bar_colors,
               alpha=0.8)
plt.xticks(range(len(ATT_performances['ARI'])),
           list(ATT_performances['ARI'].keys()),
           rotation=45)
plt.title('ARI')
plt.ylim([0, 1])
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Overall Title
plt.suptitle('Performances on ATT dataset', fontsize=16)

plt.tight_layout()
plt.savefig('figures/ATT_performances.png')
plt.show()
```
